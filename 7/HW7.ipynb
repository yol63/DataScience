```python
1.

SELECT CustomerName,Address,Country FROM Customers
WHERE City = 'Madrid' OR Country IN ('Germany','France');

2.

SELECT Country, count() as CNT FROM Customers
GROUP BY Country
ORDER BY CNT DESC
limit 3;

3.

select ShipperName,OrderDate from 
(select * from (
SELECT * FROM Orders ORDER BY OrderDate limit 10
) ORDER BY OrderDate desc limit 1) a
left join Shippers sh ON a.ShipperID = sh.ShipperID;

4.

with report_orders as 
(
SELECT *, o.Quantity * p.Price as Summ FROM OrderDetails o
left join Products p ON o.ProductID = p.ProductID 
)

select ProductName,Quantity,Price,Summ from 
(SELECT OrderID, SUM(Summ) as summm from
report_orders
GROUP BY OrderID ORDER BY summm desc limit 1) a
left join report_orders r ON a.OrderID = r.OrderID;

5.

with report_order as (SELECT ProductID, SUM(Quantity) as cnt from
OrderDetails
GROUP BY ProductID ORDER BY cnt desc limit 1)

select ProductName,Quantity,OrderID 
from report_order a
left join OrderDetails r ON a.ProductID = r.ProductID
left join Products p ON p.ProductID = a.ProductID;

6.

with report as (select * from
OrderDetails o
left join Products p ON p.ProductID = o.ProductID
left join Suppliers s ON s.SupplierID = p.SupplierID)

select SupplierName,Country,ContactName,Phone from (
select SupplierID, count(DISTINCT OrderID) as CNT 
from report
GROUP BY SupplierID 
ORDER BY CNT desc
limit 5
) p
left join Suppliers s ON s.SupplierID = p.SupplierID;

7.

with orderss as (SELECT *, p.Price * od.Quantity as Summ FROM Orders o
left join Customers c ON o.CustomerID = c.CustomerID
left join OrderDetails od ON od.OrderID = o.OrderID
left join Products p ON p.ProductID = od.ProductID
left join Categories cat ON cat.CategoryID = p.CategoryID
WHERE Country = 'Brazil')

select Country,CategoryName, SUM(Summ) as summm
from orderss 
GROUP BY Country,CategoryName
ORDER BY Summm desc
limit 1;

8.

SELECT MAX(Summ) - MIN(Summ)
FROM 
(SELECT OrderID, SUM(p.Price * od.Quantity) as Summ
FROM OrderDetails od
LEFT JOIN Products p ON p.ProductID = od.ProductID
LEFT JOIN Suppliers s ON p.SupplierID = s.SupplierID
WHERE Country == 'USA'
GROUP BY OrderID)
ORDER BY OrderID;

9.

select COUNT(DISTINCT OrderID) as CNT, a.LastName ||' '||a.FirstName as Name from 
(SELECT * FROM Employees
ORDER BY BirthDate desc
limit 3) a
LEFT JOIN Orders o ON a.EmployeeID = o.EmployeeID
GROUP BY a.EmployeeID

10.

SELECT SUM(Quantity) FROM
(SELECT * FROM [Products]
WHERE ProductName = 'Boston Crab Meat') a
LEFT JOIN OrderDetails od ON a.ProductID = od.ProductID;
```
